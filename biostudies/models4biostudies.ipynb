```python
from pandas import read_json
```

```python
df = read_json('https://www.cancermodels.org/api/search_index?select=external_model_id,data_source,project_name,model_type,histology,dataset_available,model_images,scores')
df = df[df['model_type'] != 'other']
df['mcs'] = [score ['pdx_metadata_score'] if score['pdx_metadata_score'] > score['in_vitro_metadata_score'] else score['in_vitro_metadata_score'] for score in df['scores']]
df['data_score'] = [score['data_score'] for score in df['scores']]

pdx = df[df['model_type'] == 'PDX']
in_vitro = df[df['model_type'] != 'PDX']
pdx['dataset_available'].fillna('', inplace=True)
in_vitro['dataset_available'].fillna('', inplace=True)
```

```python
in_vitro.sort_values(['mcs', 'data_score'], ascending=[False, False])
```

```python
pdx.sort_values(['mcs', 'data_score'], ascending=[False, False])
```
